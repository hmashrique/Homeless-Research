## Yearwise comparison of Male Female homelessness

```python
sheets=['2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']

file_path = "datasets/2007-2023-PIT-Counts-by-CoC.xlsb"

xls = pd.ExcelFile(file_path)

results = []
race_results = []

for sheetname in sheets:
    
    df = xls.parse(sheetname)  # Read each sheet into a DataFrame
    
    male_female = df[['Overall Homeless - Female','Overall Homeless - Male']].sum()
    
    agegroup = df[['Overall Homeless - White','Overall Homeless - Black, African American, or African',
            'Overall Homeless - Asian or Asian American','Overall Homeless - American Indian, Alaska Native, or Indigenous',
            'Overall Homeless - Native Hawaiian or Other Pacific Islander','Overall Homeless - Multiple Races']].sum()

    # Convert Series to DataFrame with features as columns
    new_df = pd.DataFrame([male_female], columns = male_female.index)
    new_df['year'] = sheetname
    
    race_df = pd.DataFrame([agegroup], columns = agegroup.index)
    race_df['year'] = sheetname
    
    race_results.append(race_df)
    results.append(new_df)

final_df = pd.concat(results, ignore_index=True)
final_race = pd.concat(race_results, ignore_index=True)

print(final_df)    
    
    
```

## Yearwise comparison of Male Vs Female homelessness

```python
# Reshape the data to a long format for easier plotting
final_df_melted = final_df.melt(id_vars='year', 
                                value_vars=['Overall Homeless - Female', 'Overall Homeless - Male'], 
                                var_name='Category', 
                                value_name='Count')
# Plot using Seaborn
sns.barplot(data=final_df_melted, x='year', y='Count', hue='Category', palette='bright')
```

## Analyzing trend year by year:

```python
# Line Plot using Seaborn
sns.lineplot(data=final_df_melted, x='year', y='Count', hue='Category', palette='bright')
```

The homelessness increased sharply from 2014 to 2015, but from 2015 till 2020, homelessness was in a stable situation with marginal increases. Then there was a dip in 2021 that maybe due to the covid 19 pandemic. From 2022, homeless again began to increase reaching the same situation of 2020. Alaramingly, in 2023, homelessness increase surpassed all previous numbers with a sharp increase similar to 2014-2015.

### There are no unaccompanid female(below 25) data from 2008-2015.

```python
unaccompanied_female = []

for sheetname in unaccompanied_sheets:
    
    df = xls.parse(sheetname)  # Read each sheet into a DataFrame
    
    # # converting the column to numeric to avoid the existing string data 
    # df["Unsheltered Homeless - Female"] = pd.to_numeric(df["Unsheltered Homeless - Female"], errors="coerce")

    female_nocompany = df[['Overall Homeless Unaccompanied Youth (Under 25) - Female']].sum()/2
    
    # Convert Series to DataFrame with features as columns
    under25_female = pd.DataFrame([female_nocompany], columns = female_nocompany.index)
    under25_female['year'] = sheetname
    
    unaccompanied_female.append(under25_female)

final_female = pd.concat(unaccompanied_female, ignore_index=True)

print(final_female)    
```

In 2021, there is a hug decline in homelessness count which maybe due to the covid-19 pandemic.
