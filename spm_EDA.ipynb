## SPM data analysis: 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read Data:

import pandas as pd

sheet = "2021"
#df = pd.read_excel(file_path, sheet_name=sheet_name)

spm= pd.read_excel("datasets/System-Performance-Measures-Data.xlsx", sheet_name = sheet)

spm.head()
```

```python
# Drop the header row
spm.columns = spm.iloc[0]  # Set the first row as header

spm = spm[1:].reset_index(drop=True)  # Drop the first row and reset the index

spm.head()
```

```python
spm.columns = [
    f"Unnamed_{i}" if pd.isna(col) else col
    for i, col in enumerate(spm.columns, start=1)
]
```

```python
spm.head()
```

```python
# rename columns

# Renaming specific columns
rename_dict = {
    "Unnamed_1": "State",
    "Unnamed_2": "CoC_Name",
    "Unnamed_3": "CoC Number",
    "Unnamed_4": "CoC_Award_Amount",
    "Unnamed_5": "CoC_Category"
}
spm.rename(columns=rename_dict, inplace=True)

spm.head()
```
